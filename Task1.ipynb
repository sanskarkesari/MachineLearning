```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print(df)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(y= df['Scores'], x= df['Hours'])
plt.title('Score Vs Study Hours',size=15)
plt.ylabel('Percentage', size=11)
plt.xlabel('Hours', size=11)
plt.show()
```

```python
sns.regplot(x= df['Hours'], y= df['Scores'])
plt.title('Regression Graph',size=15)
plt.ylabel('Percentage', size=12)
plt.xlabel('Hours', size=12)
plt.show()

```

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("TRAINED")
```

```python
pred_a = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted_Scores': [k for k in pred_a]})
prediction
```

```python
plt.scatter(x=val_X, y=val_y, color='red')
plt.plot(val_X, pred_a, color='Black')
plt.title('Actual vs Predicted', size=15)
plt.ylabel('Percentage', size=11)
plt.xlabel('Hours', size=11)
plt.show()
```

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(answer[0]))
```
